```python
# import all necessary plugins
import matplotlib.pyplot as plt
import numpy as np

# function that imports the data obtained from cpp algo
def readFile(fileName):
        fileObj = open(fileName, 'r') #opens the file in read mode
        words = fileObj.read().splitlines() #puts the file into an array
        fileObj.close()
        return words
#now we get the data and store it in rawData then make that our x-axis    
rawData = readFile("programdata_bin.txt")
xData = rawData

#turn the text into a float
for count in range(10000):
    xData[count] = float(xData[count])

#turn list into array
xData = np.array(xData)
lenData = xData.size
#maxData = np.max(xData)
#minData = np.min(xData)


xData = np.sort(xData)
cdf = np.arange(0, lenData, 1)
cdf = cdf/lenData


tempF = (1-cdf)
pdf = [10000]
for count in range(10001):
    pdf.append(1-cdf[count])

#normalize pdf
np.pdf = tempF/np.sum(tempF)

#found from trial and error
A= 200
pdfTheo = np.exp(-A*xData)

#normalize pdfTheo
pdfTheo = pdfTheo/np.sum(pdfTheo)

cdfTheo = 1 - np.exp(-A*xData)

#find error between calculated energy and theoretical
cdfErr = np.abs(cdf - cdfTheo)
pdfErr = np.abs(pdf - pdfTheo)


#check that all numbers are correct
print (lenData)
print (xData.size)
print (cdf.size)
print (np.sum(pdf))
print(np.sum(pdfTheo))
print ("sum of pdf Err: ", np.sum(pdfErr))

#just change the condition when you want to plot one
if (0 == 1):
    #p1, = plt.plot(xData, cdf, 'b', label='calculated')
    p2, = plt.plot(xData, cdfTheo, 'r+', label='theoretical') 
    plt.grid()
    plt.ylabel('probability')
    plt.xlabel('E of particle')
    plt.title('Cumulative Distribution Function (CDF)')
elif (1 == 1):
    #p1, = plt.plot(xData, pdf, 'b', label='calculated')
    p2, = plt.plot(xData, pdfTheo, 'r+', label='theoretical') 
    plt.grid()
    plt.ylabel('probability')
    plt.xlabel('E of particle')
    plt.title('Probability Density Function (pdf)')
    #plt.ylim([0,0.0002])

else:
    p1, = plt.plot(xData, pdfErr, 'b', label='pdfErr')
    p2, = plt.plot(xData, cdfErr, 'r', label='cdfErr') #when comparing data change r+ to ro, 
                                                        #it will show where it overlaps
    plt.grid()
    plt.ylabel('error')
    plt.xlabel('E of particle')
    plt.title('pdf Error')
    #plt.ylim([0,0.0002])
    

#plt.xlim([0,lenData])
#plt.ylim([minData, maxData])
#plt.legend(handles=[p1, p2], loc='upper right')

plt.show()

meanE = np.sum(pdf*xData)
print (meanE)
```
